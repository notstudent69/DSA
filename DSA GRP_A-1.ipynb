```python
class HashingDemo:
    def __init__(self, size):
        self.size = size
        self.HashTable = list(0 for _ in range(self.size))
        self.num_of_elements = 0
        self.comparisons = 0

    def isTableFull(self):
        return self.num_of_elements == self.size

    def HashFun(self, element):
        return element % self.size

    def insertElement_Linear(self, element):
        if self.isTableFull():
            print("Hash Table Full")
            return False

        position = self.HashFun(element)

        if self.HashTable[position] == 0:
            self.HashTable[position] = element
            print(f"Telephone Number {element} at position {position}")
            self.num_of_elements += 1
        else:
            print(f"Collision has occurred for Telephone number {element} at index {position}")
            position = self.linearProbing(element, position)
            self.HashTable[position] = element
            self.num_of_elements += 1
        return True

    def linearProbing(self, element, position):
        while self.HashTable[position] != 0:
            position = (position + 1) % self.size
        return position

    def insertElement_Quadratic(self, element):
        if self.isTableFull():
            print("Hash Table Full")
            return False

        position = self.HashFun(element)
        if self.HashTable[position] == 0:
            self.HashTable[position] = element
            print(f"Telephone Number {element} at position {position}")
            self.num_of_elements += 1
        else:
            print(f"Collision has occurred for Telephone number {element} at index {position}")
            occupiedStatus, position = self.quadraticProbing(element, position)
            if occupiedStatus:
                self.HashTable[position] = element
                self.num_of_elements += 1
        return True

    def quadraticProbing(self, element, position):
        limit = 50
        i = 1
        while i <= limit:
            newPosition = (position + (i ** 2)) % self.size
            if self.HashTable[newPosition] == 0:
                return True, newPosition
            i += 1
        return False, position

    def search(self, element):
        position = self.HashFun(element)
        self.comparisons += 1

        if self.HashTable[position] == element:
            return position

        temp = position - 1
        while position < self.size:
            if self.HashTable[position] != element:
                position += 1
                self.comparisons += 1
            else:
                return position

        position = temp
        while position >= 0:
            if self.HashTable[position] != element:
                position -= 1
                self.comparisons += 1
            else:
                return position

        print("Element not found")
        return False

    def display(self):
        print("\n")
        print("---------------------------")
        print("\n Position \t Telephone Number\n")
        print("---------------------------")
        for i in range(self.size):
            print(f"\t{i} ====> {self.HashTable[i]}")

hash_table_size = int(input("Enter the size of the hash table: "))
hash_object1 = HashingDemo(hash_table_size)
print("\nInserting The Telephone Numbers in the Hash Table....\n")
print("\nCollision Resolution using Linear Probing\n")

telephone_numbers = [int(input("Enter a telephone number: ")) for _ in range(10)]

for number in telephone_numbers:
    hash_object1.insertElement_Linear(number)

hash_object1.display()
print()

search_numbers = [int(input("Enter a number to search: ")) for _ in range(10)]

for number in search_numbers:
    position = hash_object1.search(number)
    if position is not False:
        print("The position of number", number, "is:", position)


print("\nTotal number of comparisons done for searching =", hash_object1.comparisons)


hash_object2 = (hash_table_size)
print("\nInserting The Telephone Numbers in the Hash Table....\n")
print("\nCollision Resolution using Quadratic Probing\n")

telephone_numbers = [int(input("Enter a telephone number: ")) for _ in range(10)]

for number in telephone_numbers:
    hash_object2.insertElement_Quadratic(number)

hash_object2.display()
print()

search_numbers = [int(input("Enter a number to search: ")) for _ in range(10)]

for number in search_numbers:
    position = hash_object2.search(number)
    if position is not False:
        print("The position of number", number, "is:", position)


print("\nTotal number of comparisons done for searching =", hash_object2.comparisons)


```
